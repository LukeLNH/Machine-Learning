```python
class neural_net_one_hidden:
    def __init__(self):
        pass 

    def back_propogation(iterations, learning_rate, targets, smoothing_parameter, num_of_hidden_nodes, num_of_output_nodes):
        
        np.random.seed(1)
        # Dimensions 
        N, D = X.shape
        M = num_of_hidden_nodes
        K = num_of_output_nodes
        
        # Inital Weights
        w = np.random.randn(D * M).reshape(D, M)
        v = np.random.randn(M * K).reshape(M, K)
        
        # Initial Biases 
        b_1 = np.random.randn(M).reshape(1, M)
        ones = np.ones(N).reshape(N, 1)
        b = np.dot(ones, b_1).astype(object)
        c_1 = np.random.randn(K).reshape(1, K)
        c = np.dot(ones, c_1).astype(object)
        
        #Targets after one hot encoding
        zeros = np.zeros(N*K).reshape(N,K)
        T = np.hstack((targets,zeros))
        
        for i in range(N):
            index_value = T[i,0]
            T[i, int(index_value)] = 1
        T = np.delete(T, 0, axis = 1)
        
        # Gradient descent loop 
        for i in range(iterations):
            # Feedforward
            Z = (1)/(1 + np.exp(-(np.dot(X.astype(float), w.astype(float)) + b.astype(float))))
            Y = np.dot(Z.astype(float), v.astype(float)) + c.astype(float)
            S, J = Y.shape
            for i in range(S):
                Y[i, :] = np.exp(Y[i, :])/(np.exp(Y[i, :]).sum())

            # Gradients 
            partial_v = Z.T.dot(T - Y)
            first = np.dot((T - Y), v.T)
            Second = np.dot(first.T, Z)
            Third = np.dot(Second, (1 - Z).T)
            partial_w = np.dot(X.T, first * Z * (1-Z))
            partial_c = (T - Y).sum(axis=0)
            partial_b = ((T - Y).dot(v.T) * Z * (1 - Z)).sum(axis=0)
            
            
            # Cost 
            total = -np.dot(T.T, np.log(Y))
            cost = total.sum()
            
            #Gradient Descent 
            v += learning_rate * (partial_v.astype(float))# - smoothing_parameter*v)
            w += learning_rate * (partial_w.astype(float))# - smoothing_parameter * w)
            b += learning_rate * partial_b.astype(float)
            c += learning_rate * partial_c.astype(float)
            
            
            # Accuracy 
        
        #Accuracy = np.mean(np.round(Y) == T) #THis way of doing it yields the same result
        
        n_correct = 0
        n_total = 0
        for i in range(len(Y)):
            for j in range(len(Y[0])):
                n_total += 1
                if np.round(Y[i,j]) == T[i,j]:
                    n_correct += 1
        Accuracy = float(n_correct) / float(n_total)
        
        #print("The weights after {} iterations of gradient descent are w = {} \n v = {}.".format(iterations, w, v))
        #print("The biases after {} iterations of gradient descent are b = {} \n c = {}.".format(iterations, b, c))
        #print ("{}% of the targets match the predictions".format(Accuracy * 100))
        print(((T - Y).dot(v.T)).shape, Z.shape)
        return type(Z), type(T), type(Y), type(v), ((T - Y).dot(v.T)).shape, Z.shape
    
        
```
