```python
# importing necessary libraries
import numpy as np
import time
import random

#defining the different game difficulty options. The player will be able to select 4 different options
def game_difficulty(difficulty):
    
    global list_case
    global index
    global ans
    
    if difficulty == 1:
        a = np.random.randint(0,10)
        b = np.random.randint(0,10)
        list_case = [[f"{a} + {b}", f"{a} - {b}"],[a + b, a - b]]
        index = random.choice([0,1])
        ans = list_case[1][index]
        print(list_case[0][index])
    
    if difficulty == 2:
        b = np.random.randint(1,10)
        a = b * (np.random.randint(0,10))
        list_case = [[f"{a} * {b}", f"{a} / {b}"],[a * b, a / b]]
        index = random.choice([0,1])
        ans = list_case[1][index]
        print(list_case[0][index])
        
    if difficulty == 3:
        operation = random.choice(["add","multiply"])
        if operation == "add":
            a = np.random.randint(0,10)
            b = np.random.randint(0,10)
            list_case = [[f"{a} + {b}", f"{a} - {b}"],[a + b, a - b]]
            index = random.choice([0,1])
            ans = list_case[1][index]
            print(list_case[0][index])
        else:
            b = np.random.randint(1,10)
            a = b * (np.random.randint(0,10))
            list_case = [[f"{a} * {b}", f"{a} / {b}"],[a * b, a / b]]
            index = random.choice([0,1])
            ans = list_case[1][index]
            print(list_case[0][index])
            
    if difficulty == 4:
        operation = random.choice(["add","multiply"])
        if operation == "add":
            a = np.random.randint(0,100)
            b = np.random.randint(0,100)
            list_case = [[f"{a} + {b}", f"{a} - {b}"],[a + b, a - b]]
            index = random.choice([0,1])
            ans = list_case[1][index]
            print(list_case[0][index])
        else:
            b = np.random.randint(1,10)
            a = b * (np.random.randint(0,100))
            list_case = [[f"{a} * {b}", f"{a} / {b}"],[a * b, a / b]]
            index = random.choice([0,1])
            ans = list_case[1][index]
            print(list_case[0][index])    


#defining variables used to evaluate the player's performance at the end of the game
score_correct = 0
score_incorrect = 0
streak_num = 0
streak_list = []

#loop to activate the game
print("Welcome to Quick Math! Answer as many questions as you can in 30 seconds!") 
print("Would you like to start playing? Answer Yes/No")
player_descision = input()
if player_descision.lower() == "yes":
    game_on = True
elif player_descision.lower() == "no":
    game_on = False
    print("Come and play again next time!")
else:
    print("You entered neither Yes or No. The program will now terminate. :(")
    game_on = False


#the game has been activated
while game_on == True:
    
    #loop to select the game mode and difficulty level
    print("Select the operation you want to practice, 1(addition), 2(multiplication), 3(mixed, basic) or 4 (mixed, advanced)")
    difficulty = input()
    if float(difficulty) == 1:
        print("You have chosen to practice basic addition and subtraction!")
    elif float(difficulty) == 2:
        print("You have chosen to practice basic multiplication and division!")
    elif float(difficulty) == 3:
        print("you have chosen to practice all four operations at the basic level! Excellent!")
    elif float(difficulty) == 4:
        print("you have chosen to practice all four operations at the advanced level! Way to go!")
    else:
        print("The computer could not understand what you want to practice. The program will now terminate. :(")
        game_on = False
        break
    time.sleep(1)
    
    #precursor to game loop i.e time for player to get ready
    print("the Timer will start in:")
    for i in [3,2,1]:
        print(i)
        time.sleep(1)
    stop_time = time.perf_counter() + 30
    
    #game loop
    while time.perf_counter() < stop_time:
        game_difficulty(int(difficulty))
        player_guess = input()
        if int(player_guess) == ans:
            print("Good Job! :D")
            score_correct += 1
            streak_num += 1
        else:
            print("You're wrong :(")
            score_incorrect += 1
            streak_list.append(streak_num)
            streak_num = 0
            
    #Evaluating the player
    print(f"Number of problems you attempted: {score_correct + score_incorrect}")
    print(f"Number of problems you answered correctly: {score_correct}")
    print(f"Your Highest Streak: {max(streak_list)}")
    if score_correct >= 20 and float(difficulty) == 4:
        print("You have practically mastered mental math!")
    elif score_correct >= 20:
        print("You should move on to the next level :)")
        
    #asking to play the game again
    print("Would you like to play again? Yes or No? :)")
    continue_choice = input()
    if continue_choice.lower() == "yes":
        game_on = True
    elif continue_choice.lower() == "no":
        print("Aww okay. Looking forward to your next visit! :)")
        game_on = False
    else:
        print("You entered neither Yes or No. The program will now terminate. :(")
        game_on = False
```
