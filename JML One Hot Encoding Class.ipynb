```python
import numpy as np
import pandas as pd
```

```python
class One_Hot_Encoder_FromData():
    def __init__(self,data): #data must be passed in as str
        mat = np.asmatrix(pd.read_csv(data))
        self.mat = mat
    
    def ohe(self, column_index):
        num_of_categories = max(self.mat[:,column_index])
        N,D = self.mat.shape
        zeros = np.zeros(N*int(num_of_categories)).reshape(N,int(num_of_categories))
        ohe_mat = np.hstack((self.mat,zeros))
        
        for num in range(N):
            t = int(ohe_mat[num,column_index])
            ohe_mat[num,D-1+t] = 1
        return ohe_mat
```

```python
class One_Hot_Encoder_FromMat():
    def __init__(self,mat): #data must be passed in as str
        self.mat = mat
    
    def ohe(self, column_index):
        num_of_categories = max(self.mat[:,column_index])
        N,D = self.mat.shape
        zeros = np.zeros(N*int(num_of_categories)).reshape(N,int(num_of_categories))
        ohe_mat = np.hstack((self.mat,zeros))
        
        for num in range(N):
            t = int(ohe_mat[num,column_index])
            ohe_mat[num,D-1+t] = 1
        return ohe_mat
```

```python
class One_Hot_Encoder_FromMat_V2():
    def __init__(self,mat):
        self.mat = mat
    
    def ohe(self, column_index):
        num_of_categories = max(self.mat[:,column_index])
        N,D = self.mat.shape
        zeros = np.zeros(N*int(num_of_categories)).reshape(N,int(num_of_categories))
        ohe_mat = np.hstack((self.mat,zeros))
        
        ohe_mat[np.arange(N),D - 1 + ohe_mat[:,column_index].astype(np.int)] = 1
        return self.mat
```

```python
class Standardization():
    def __init__(self, mat):
        self.mat = mat
        print(mat)
    def Standardize_Values(self,*column):
        matt = self.mat
        matt[:,column] = (matt[:,column] - matt[:,column].mean())/matt[:,column].std()
        return matt
```

```python
movie_ohe_mat = Standardization(movie_ohe_mat)
```

```python
movie_ohe_mat = movie_ohe_mat.Standardize_Values(0,1)
```
